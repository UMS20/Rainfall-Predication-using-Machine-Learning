2. Write a function to Calculate Area and Perimeter of Rectangle

```python
def rectangle_properties(length, width):
    area = length * width
    perimeter = 2 * (length + width)
    return area, perimeter

# Test the function
length = float(input("Enter the length of the rectangle: "))
width = float(input("Enter the width of the rectangle: "))

area, perimeter = rectangle_properties(length, width)
print("Area of the rectangle:", area)
print("Perimeter of the rectangle:", perimeter)
```

3. Write a python Program to find out given No is Prime or Not.

```python
def is_prime(number):
    if number <= 1:
        return False
    elif number <= 3:
        return True
    elif number % 2 == 0 or number % 3 == 0:
        return False
    i = 5
    while i * i <= number:
        if number % i == 0 or number % (i + 2) == 0:
            return False
        i += 6
    return True

# Test the function
number = int(input("Enter a number: "))

if is_prime(number):
    print(number, "is a prime number.")
else:
    print(number, "is not a prime number.")
```

4. Write a Python program that prompts the user to input three numbers and then determines and
prints the largest among them.

```python
def find_largest(num1, num2, num3):
    return max(num1, num2, num3)

# Prompt the user to input three numbers
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))
num3 = float(input("Enter the third number: "))

# Call the function to find the largest number
largest_number = find_largest(num1, num2, num3)

# Print the result
print("The largest number is:", largest_number)
```

5. Write a Python program to calculate electricity bill (accept no of unit from user) according to following criteria: Unit Price
First 100 units  no charge
Next 100 units Rs 5 per Unit
After 200 Units Rs 10 per Unit
(For Example if input unit is 350 then total bill amount is 2000)

```python
def calculate_bill(units):
    if units <= 100:
        bill_amount = 0
    elif units <= 200:
        bill_amount = (units - 100) * 5
    else:
        bill_amount = (100 * 5) + (units - 200) * 10
    return bill_amount

# Prompt the user to input the number of units consumed
units_consumed = int(input("Enter the number of units consumed: "))

# Calculate the bill amount
total_bill = calculate_bill(units_consumed)

# Print the total bill amount
print("Total electricity bill amount:", total_bill)
```

6. Write a program to accept numbers from 1-7 and display name of day like- 1 for sunday, 2 for Monday and so on.

```python
def get_day_name(number):
    days = {
        1: "Sunday",
        2: "Monday",
        3: "Tuesday",
        4: "Wednesday",
        5: "Thursday",
        6: "Friday",
        7: "Saturday"
    }
    return days.get(number, "Invalid day number")

# Prompt the user to input a number from 1 to 7
number = int(input("Enter a number from 1 to 7: "))

# Get the day name and print it
day_name = get_day_name(number)
print("The name of the day is:", day_name)
```

7. Write a python program to check given word is Palindrome or not

```python
def is_palindrome(word):
    # Convert the word to lowercase to handle case-insensitivity
    word = word.lower()
    # Compare the word with its reverse
    return word == word[::-1]

# Prompt the user to input a word
word = input("Enter a word: ")

# Check if the word is a palindrome and print the result
if is_palindrome(word):
    print("The word is a palindrome.")
else:
    print("The word is not a palindrome.")
```
